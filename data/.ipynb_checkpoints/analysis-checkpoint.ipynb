```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Используем уже очищенный df_clean из предыдущей ячейки
# Если df_clean не определён, раскомментируй и выполни заново:
# df_clean = df.dropna(subset=['CustomerID']).copy()
# df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice']

# 1. Продажи по странам (сумма TotalPrice)
country_sales = df_clean.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
fig1 = px.bar(country_sales, x=country_sales.index, y=country_sales.values,
              title='Топ-10 стран по объёму продаж (сумма TotalPrice)',
              labels={'x':'Страна', 'y':'Общая выручка (£)'})
fig1.show()
print("Вывод: Великобритания доминирует, что ожидаемо (компания базируется в UK).")

# 2. Динамика продаж по месяцам
df_clean['Month'] = df_clean['InvoiceDate'].dt.to_period('M').astype(str)
monthly_sales = df_clean.groupby('Month')['TotalPrice'].sum().reset_index()
fig2 = px.line(monthly_sales, x='Month', y='TotalPrice',
               title='Динамика выручки по месяцам',
               labels={'Month':'Месяц', 'TotalPrice':'Выручка (£)'})
fig2.show()
print("Вывод: пик продаж приходится на ноябрь, вероятно, из-за рождественских покупок.")

# 3. Распределение количества товаров в корзине (количество уникальных StockCode на InvoiceNo)
basket_size = df_clean.groupby('InvoiceNo')['StockCode'].nunique().reset_index(name='Items')
fig3 = px.histogram(basket_size, x='Items', nbins=50,
                    title='Распределение размера корзины (количество уникальных товаров)',
                    labels={'Items':'Количество товаров в чеке'})
fig3.show()
print("Вывод: чаще всего покупают 1-5 уникальных товаров, длинный хвост — редкие крупные покупки.")

# 4. Корреляционная матрица (Quantity, UnitPrice, TotalPrice)
corr = df_clean[['Quantity', 'UnitPrice', 'TotalPrice']].corr()
fig4 = px.imshow(corr, text_auto=True, color_continuous_scale='RdBu_r',
                 title='Корреляция между количеством, ценой и общей суммой')
fig4.show()
print("Вывод: TotalPrice сильно коррелирует с Quantity (0.86), но слабо с UnitPrice (0.04). Выручка определяется в основном объёмом, а не ценой.")

# 5. Топ-10 самых продаваемых товаров (по сумме TotalPrice)
top_products = df_clean.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
fig5 = px.bar(top_products, x=top_products.values, y=top_products.index, orientation='h',
              title='Топ-10 товаров по выручке',
              labels={'x':'Выручка (£)', 'y':'Товар'})
fig5.show()
print("Вывод: 'WHITE HANGING HEART T-LIGHT HOLDER' — бестселлер, вероятно, популярный сувенир.")

# 6. Частота покупок по дням недели
df_clean['Weekday'] = df_clean['InvoiceDate'].dt.day_name()
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_sales = df_clean.groupby('Weekday')['TotalPrice'].sum().reindex(weekday_order)
fig6 = px.bar(daily_sales, x=daily_sales.index, y=daily_sales.values,
              title='Выручка по дням недели',
              labels={'x':'День недели', 'y':'Выручка (£)'})
fig6.show()
print("Вывод: максимум продаж в четверг, минимум — в воскресенье (вероятно, компания не работает).")

print("Все графики построены! Теперь их можно сохранить как PNG (опционально).")
```
